```python
import pandas as pd
import pathlib
import re
whole_df: pd.DataFrame = pd.DataFrame()
for path in pathlib.Path(".").glob("*.csv"):
    if path.name.endswith(".csv"):
        whole_df = pd.concat([whole_df, pd.read_csv(path.name).assign(file = re.sub(r".csv$", "", str(path.name)))]).reset_index(drop = True)
whole_df.loc[:,'County'] = whole_df['County'].fillna(method = "ffill")

display(whole_df)                                                                 
```

```python
def toint(s):
    try: 
        return int(s)
    except ValueError:
        return 0
whole_df[['Final Expression', "First Expression"]] = whole_df[['Final Expression', "First Expression"]].applymap(toint)
display(whole_df)
whole_df.loc[:,'vote_diff'] = whole_df['Final Expression'].astype(int) - whole_df['First Expression'].astype(int)
whole_df.loc[:,'positive_vote_diff'] = whole_df['vote_diff'].map(lambda x: x > 0)
display(whole_df.groupby("file").agg('sum'))
```
